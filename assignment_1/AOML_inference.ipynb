```python
import joblib
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
scaler = joblib.load(r"C:\Users\khushi shetty\OneDrive\Desktop\AOML\scaler.pkl")
ohe = joblib.load(r'C:\Users\khushi shetty\OneDrive\Desktop\AOML\ohe.pkl')
knn_model = joblib.load(r'C:\Users\khushi shetty\OneDrive\Desktop\AOML\knn.pkl')
dt_model = joblib.load(r'C:\Users\khushi shetty\OneDrive\Desktop\AOML\dt_classifier.pkl')
```

```python
num_col = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']
cat_col = ['BMI_category']
def preprocess(data):
    data[num_col] = scaler.transform(data[num_col])
    data_encoded = pd.DataFrame(ohe.transform(data[cat_col]).toarray(),columns=ohe.get_feature_names_out(cat_col))
    data = data.drop(cat_col, axis=1)
    data = pd.concat((data, data_encoded), axis=1)
    
    return data 

def knn_predict(data):
    data = preprocess(data)
    return knn_model.predict(data)

def dt_predict(data):
    data = preprocess(data)
    return dt_model.predict(data)
```
